```python
# Importing Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
gt = pd.read_csv("globalterrorismdb_0718dist.csv",encoding ='latin1',low_memory=False)
gt.head()
```

```python
gt.describe()
```

```python
gt.info()
```

```python
gt.dtypes
```

```python
gt.columns.values
```

```python
gt=gt[['eventid', 'iyear', 'imonth', 'country', 'region','provstate','city','crit1', 'crit2', 'crit3','success', 'suicide', 'attacktype1','targtype1','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
gt.head()
```

```python
gt.isnull().sum()
```

```python
gt['nkill']=gt['nkill'].fillna(0)
gt['nwound']=gt['nwound'].fillna(0)
gt.isnull().sum()
```

```python
gt['casualities']=gt['nkill']+gt['nwound']
gt.describe()
```

```python
print(f"""
There are {gt.country.nunique()} Countries from {gt.region.nunique()} Regions covered in the dataset terrorist attack data in {gt.claimed.nunique()} years from {gt.iyear.min()} to {gt.iyear.max()}. Overall {gt.index.nunique()} terrorist attacks are recorded here which caused about {gt.casualities.sum()} Casualities Consisted of {gt.nkill.sum()} Kills and {gt.nwound.sum()} Wounded.
""")
```

# DATA VISUALIZATION

```python
yearc=gt[['iyear','casualities']].groupby('iyear').sum()
yearc.plot(kind='bar',color='green',figsize=(15,6))
plt.title("Casualities")
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Casualities',fontsize=15)
plt.show()
```

```python
gt['attacktype1'].value_counts().plot(kind='bar',figsize=(20,10),color='red')
plt.xlabel('Attack Type',fontsize=20)
plt.ylabel('Number of attacks',fontsize=20)
plt.title('Number of attacks',fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(15,6))
country_attacks = gt.gname.value_counts()[:15].reset_index()
country_attacks.columns = ['gname', 'Total Attacks']
sns.barplot(x = country_attacks.gname, y = country_attacks['Total Attacks'], palette='OrRd_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation = 90)
plt.title("Number of Total Attacks in Each Country")
plt.show()
```
